### Import Dataset

```python
import pandas as pd
data = pd.read_csv("grocery_chain_data.csv")
print(data.head())
```

### Basic Inspection of Dataset

```python
data.info()
data.describe()
```

### Find any NULL values in the Dataset

```python
print(data.isnull().sum())
```

### Drop any NULL value row

```python
data = data.dropna().reset_index(drop=True)
print(data.head())
```

### Check after cleaning the dataset shape

```python
data.shape
```

###  Check any Duplicate datas in Dataset

```python
print(data.duplicated().sum())
```

### Filter the data when final_amount is not in negative

```python
data= data[data['final_amount']>= 0].reset_index(drop=True)
```

### Sales by Store

```python
store_summary=data.groupby("store_name")["final_amount"].sum()
print(store_summary.sort_values(ascending=False))
```

```python
print(data["transaction_date"].dtype)
```

### The transaction_date is not in Correct  Format. So change the datatype.

```python
data["transaction_date"]=pd.to_datetime(data["transaction_date"], dayfirst=False)
print(data["transaction_date"])
print("\nDatetime column changed:",data["transaction_date"].dtype)
```

### Create  a new column in existing table.

```python
data["Year"]= data["transaction_date"].dt.year
data["Month"]=data["transaction_date"].dt.month_name()
data["Day_of_week"] = data["transaction_date"].dt.day_name()
print("Transaction year\n",data["Year"],"\nTransaction Month\n",data["Month"],"\nTransaction Day\n",data["Day_of_week"])
```

### After Create new column. we check it.

```python
print(data.head())
```

### sales by year

```python
sales_by_year = data.groupby("Year")["final_amount"].sum()
print("________Sales By Year_______\n",sales_by_year.sort_values(ascending=False))
```

### Slaes by Month

```python
sales_by_month=data.groupby("Month")["final_amount"].sum()
print("_______Sales By Month________\n",sales_by_month.sort_values(ascending=False))
```

### Sales by Day of week

```python
Week_sale = data.groupby("Day_of_week")["final_amount"].sum()
print("____Highest sales day of week____\n",Week_sale.sort_values(ascending=False))
```

### Highest Sales by one day and based Month, day and Year

```python
Highest_rate_of_sale=data.groupby(["Year","Month","Day_of_week"])["final_amount"].sum()
print(Highest_rate_of_sale.sort_values(ascending=False))
```

### Sales by Quantity

```python
High_store_level = data.groupby("store_name")[["quantity","final_amount"]].sum()
print(High_store_level.sort_values(by="quantity",ascending=False))
```

### Which user ID Had highest Loyalty points.

```python
loyalty =data.groupby("customer_id")["loyalty_points"].sum().head()
print(loyalty.sort_values(ascending=False))
```
